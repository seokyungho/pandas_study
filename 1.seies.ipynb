#pandas
파이썬에서 분석하는 데이터 분석 라이브러리


```python
import pandas as pd
```

#1.series
1차원 데이터 (정수 실수 문자열 등)


ex)1월부터 4월까지 평균 온도 데이터 (-20 , -10 , 10 , 20)

```python
temp= pd.Series([-20,-10,10,20])
print(temp)
```

```python
temp[0] #1월 온도
```
