#데이터 선택(loc)
이름을 이용하여 원하는 row에서 원하는 col 선택

```python
df.loc['1번'] #index 1번에 해당하는 전체 데이터
```

```python
df.loc['1번','국어'] #index 1번의 국어 성적
```

```python
df.loc[['1번','2번'],['국어','영어']] #1번 2번의 국어 성적 , 영어 성적
```

```python
df.loc['1번':'5번','국어':'사회'] #index 1번부터 5번 까지 , 국어부터 사회 성적 index 정수랑로 하는 방식과 다름
```

데이터 선택 (iloc) integer location
위치를 이용하여 원하는 row 에서 원하는 col 선택

```python
df.iloc[0] #0번째 위치의 데이터 가져옴
```

```python
df.iloc[0:5] # 0~4번째 위치의 데이터 
```

```python
df.iloc[0,1] #0번째 위치의 1번째 데이터
```

```python
df.iloc[[0,1],2] #0,1 번째의 2번째 데이터
```

```python
df.iloc[[0,1],[4,5]] #0,1번째 위치의 4,5 번째 데이터
```

```python
df.iloc[0:5 , 3:8] #슬라이싱엔 따로 대괄호 안해줘도 됨
```

데이터 선택 (조건)
조건에 해당하는 데이터 가져오기

```python
df['키'] >= 185
```

```python
filt = (df['키'] >= 185)
df[filt]
```

```python
df[~filt] # ~을 앞에 붙히면 역필터
```

```python
df.loc[filt,'수학']
```

```python
df.loc[filt,['수학','과학','사회']]
```

다양한 조건

&

```python
df.loc[(df['키']>= 185) & (df['학교']=='북산고')]
```

or |

```python
df.loc[(df['키']<170)|(df['키']>200)]
```
